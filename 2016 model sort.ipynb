```python
#Yearly dataset, import from xlsx
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
csv_path= "2016-sales.csv"
sales_df = pd.read_csv(csv_path)
sales_df.head(42)
```

```python
sales_df.rename(columns={"2016-US":"Model"}, inplace=True)
sales_df['JAN'] = pd.to_numeric(sales_df.JAN.astype(str).str.replace(',',''), errors='coerce', downcast='integer')
sales_df['FEB'] = pd.to_numeric(sales_df.FEB.astype(str).str.replace(',',''), errors='coerce', downcast='integer')
sales_df['MAR'] = pd.to_numeric(sales_df.MAR.astype(str).str.replace(',',''), errors='coerce', downcast='integer')
sales_df['APR'] = pd.to_numeric(sales_df.APR.astype(str).str.replace(',',''), errors='coerce', downcast='integer')
sales_df['MAY'] = pd.to_numeric(sales_df.MAY.astype(str).str.replace(',',''), errors='coerce', downcast='integer')
sales_df['JUN'] = pd.to_numeric(sales_df.JUN.astype(str).str.replace(',',''), errors='coerce', downcast='integer')
sales_df['JUL'] = pd.to_numeric(sales_df.JUL.astype(str).str.replace(',',''), errors='coerce', downcast='integer')
sales_df['AUG'] = pd.to_numeric(sales_df.AUG.astype(str).str.replace(',',''), errors='coerce', downcast='integer')
sales_df['SEP'] = pd.to_numeric(sales_df.SEP.astype(str).str.replace(',',''), errors='coerce', downcast='integer')
sales_df['OCT'] = pd.to_numeric(sales_df.OCT.astype(str).str.replace(',',''), errors='coerce', downcast='integer')
sales_df['NOV'] = pd.to_numeric(sales_df.NOV.astype(str).str.replace(',',''), errors='coerce', downcast='integer')
sales_df['DEC'] = pd.to_numeric(sales_df.DEC.astype(str).str.replace(',',''), errors='coerce', downcast='integer')
sales_df['Total'] = pd.to_numeric(sales_df.Total.astype(str).str.replace(',',''), errors='coerce', downcast='integer')
sales_df.fillna(value=0, inplace=True)
sales_df.head(50)
```

```python
clean_df = sales_df.sort_values(by=['Total'], ascending=False)
clean_df.head(50)
```

```python
topten_df= clean_df.iloc[0:10]
topten_df.head(10)
```

```python
totalEV=clean_df["Total"].sum()
modelS=28896
volt=24139
modelX=18223
fusion=15938
leaf=14006
top5= (modelS+volt+modelX+fusion+leaf)
others=totalEV-top5
totalEV
```

```python
labels = ["Tela Model S", "Chevrolet Volt", "Tesla Model X", "Ford Fusion Energi", "Nissan LEAF", "All Others"]
qtys=[modelS, volt, modelX, fusion, leaf, others ]
colors = ["red", "blue", 'green', 'yellow', 'purple', 'orange']
explode = (.1, 0,0,0,0,0)
plt.pie(qtys, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=135)
plt.axis("equal")
plt.title("US EV Market")
plt.xlabel("Top 5 Models")
plt.show()
plt.savefig("2016-EV-pie.png")
```
